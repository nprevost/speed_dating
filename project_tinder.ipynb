**Pour obtenir un deuxième rendez-vous, est-il préférable d'être le premier speed date de la soirée de quelqu'un ou le dernier ?**

```python
#Mask pour un match des 2 cotes
df_second_date = df_speed_dating_clean[(df_speed_dating_clean['dec'] == 'Y') & (df_speed_dating_clean['dec_o'] == 'Y')]
df_second_date.shape
```

Duplication pour avoir un dataframe avec juste 1 personne sur les 2

```python
def deduplicate_df(df):
    """
    Deduplicates rows in a DataFrame based on iid and pid relationships.

    Args:
        df: The input DataFrame.

    Returns:
        A new DataFrame with duplicate rows removed.
    """

    # Create a set to store processed pairs
    processed_pairs = set()
    deduplicated_df = pd.DataFrame()

    for index, row in df.iterrows():
      iid = row['iid']
      pid = row['pid']

      # Check for reverse pair in processed_pairs
      if (pid, iid) in processed_pairs:
          continue
      else:
        deduplicated_df = pd.concat([deduplicated_df, pd.DataFrame([row])], ignore_index=True)
        processed_pairs.add((iid, pid))

    return deduplicated_df

# Deduplicate the DataFrame
df_second_date_deduplicated = deduplicate_df(df_second_date)

# Display or further process the deduplicated DataFrame
df_second_date_deduplicated.shape
```

```python
# Renommage des colonnes pour indiquer la 2eme personne
for column in df_second_date_deduplicated.columns:
  df_second_date_deduplicated.rename(columns={column: column + "_2"}, inplace = True)

df_second_date_deduplicated.columns
```

```python
# Merge des 2 dataset en gardant juste les lignes avec les 2 dataset remplis
df_second_date_test = df_second_date.merge(df_second_date_deduplicated, how = 'inner', left_on = ['iid', 'pid'], right_on = ['pid_2', 'iid_2'])
df_second_date_test
```

```python
fig = px.scatter(df_second_date_test, x = 'position', y = 'round', width = 600, color = 'gender')
fig.show()
```
